# Predicting Watershed Discharge

This notebook walks through the creation of a simple model for stream discharge from a confined catchment. For more detail and theoretical background, see "Predicting Watershed Discharge.pdf" in the parent directory.

## 1. Discrete Convolution Integral for Discharge

The discrete convolution integral 

\begin{equation}
	Q_t = \frac{A}{\tau_D} \sum_{u = 0}^{t} P_u e^{-(t-u)/\tau_D} .
\end{equation}

is used to model discharge, $Q$, following precipitation events, where the precipitation input, $P$, is a discrete time series of instantaneous, pulse inputs. The summation is limited by the bounds of the incrementing variable, $u = [0,t]$. The resulting discharge time series approximates a discharge hydrograph beginning at time $0$ and ending at time $t$. $A$ is watershed surface area and $\tau_D$ is the drainage timescale, a theoretical lumped parameter defined

\begin{equation}
    \tau_D = \frac{A \phi \beta}{K}
\end{equation}

where $\beta$ is a watershed shape parameter, $\phi$ is average porosity and $K$ is average hydraulic conductivity of the watershed's subsurface material. Precipitation inputs into the model are scaled by a fraction to account evapotranspiration water loss. For temperate, humid climates (ex. Princess Anne, MD), a good estimate is 60% lost to humidity. The other parameters are based on USGS data.

```python
A = 12.4*10**6  #m^2
phi = 0.425    #-
beta = 1.0/16    #m^-1a
K = 1.0*3600*24 #m/day
et = 0.6
tauD = (A*phi*beta)/K
print tauD
```

Let's upload a real precipitation time series to test the model on. We also upload real discharge data. Both datasets are from Princess Anne, MD and the discharge data is from the Manokin Branch stream. The time series we use measures total daily rainfall amounts in meters. We upload the data below.

```python
realPrecip = np.loadtxt('real_precipitation_timeseries.txt')
realQ = np.loadtxt('real_discharge_timeseries.txt')
realQ = realQ*0.0283168 #convert to m^3/s
```

The discrete convolution is implemented below. It requires a precipitation time series,
`precipTimeSeries`, a drainage timescale, `tauD`, and watershed surface area, `shedA`. The resultant discharge units are in cubic meters per day. We divide this by the number of seconds in a day to find average daily discharge in m^3/s.

```python
def dischargeConvolution(realPrecip, tauD, A):
    realPrecipAdj = (1-et)*realPrecip #account for et
    modelQ = np.zeros(len(realPrecipAdj))
    for i in range(0,len(realPrecipAdj)):
        qSum = 0
        for j in range(0,i):
            qSum = qSum + realPrecipAdj[i-j]*np.exp(-j/tauD)
            modelQ[i] = (A/tauD)*qSum
    return modelQ/(3600*24)
```

We plot a modelled and real hydrographs below.

```python
plt.figure(figsize=(14,4))
modelQ = dischargeConvolution(realPrecip, tauD, A)
time = np.linspace(0,len(modelQ),len(modelQ))
plt.plot(time,modelQ,
         color='tab:blue',label='Model')
plt.plot(time,realQ,color='tab:grey',label='Real')
#prettify...
plt.title('Modelled and Real Hydrograph, Manokin Branch')
plt.xlabel('Time (days)'); plt.xlim(0,len(modelQ))
plt.ylabel('Avg Daily Discharge (m$^3$/s)')
plt.legend(frameon=False)
plt.show()
```

## 2. Generating Precipitation Time Series

Precipitation can be simulated with a simple mixed distribution model. A random variable is taken from a uniform distribution. Based on the probability of a day being rainy, either a 0 value (Dirac function) or a random variable from the exponential distribution is chosen. The exponential distribution is parameterized by the mean daily rainfall intensity, $\mu_R$, which can be empirically found for any precipitation time series.

```python
precipNoZeros = np.asarray([i if i > 0 else np.nan for i in realPrecip])
mean = np.nanmean(precipNoZeros)*1000
print 'Mean daily rainfall: ' + '{:05.3f}'.format(mean) + ' mm'
```

```python
def getPrecipSeries(mean, alpha, length):
    precip = np.zeros(length)
    for i in range(0,len(precip)):
        rand = np.random.uniform()
        if rand < alpha: #rainy day
            precip[i] = np.random.exponential(mean)
        else:
            precip[i] = 0.0
    return precip
```

```python
simSeries = getPrecipSeries(mean/1000,0.33,100)
simDis = dischargeConvolution(simSeries, tauD, A)
time = np.linspace(0,len(simDis),len(simDis))

fig, ax1 = plt.subplots()
ax1.plot(time,simDis)
ax1.set_xlabel('Time (days)')
ax1.set_ylabel('Avg Daily Discharge (m$^3$/s)')
ax2 = ax1.twinx()
ax2.bar(time,simSeries,color='tab:grey',alpha=0.3)
ax2.set_ylabel('Daily Precipitation (m)')
fig = plt.gcf()
fig.set_size_inches(14, 4)
plt.show()
```
